```python
# Create the DataFrame
data = {
    'Name': ['Aparna', 'Pankaj', 'Ram', 'Ramesh', 'Naveen', 'Krrishnav', 'Bhawna'],
    'Degree': ['MBA', 'BCA', 'M.Tech', 'MBA', 'NaN', 'BCA', 'MBA'],
    'Score': [90.0, None, 80.0, 98.0, 97.0, 78.0, 89.0]
}

df = pd.DataFrame(data)

# a) To create the DataFrame
print("Original DataFrame:")
print(df)

# b) To print the Degree and maximum marks in each stream
degree_max_marks = df.groupby('Degree')['Score'].max()
print("\nMaximum marks in each stream:")
print(degree_max_marks)

# c) To fill the NaN with 76
df['Score'].fillna(76, inplace=True)
print("\nDataFrame after filling NaN with 76:")
print(df)

# d) To set the index to Name
df.set_index('Name', inplace=True)
print("\nDataFrame with index set to Name:")
print(df)

# e) To display the name and degree-wise average marks of each student
degree_avg_marks = df.groupby('Degree')['Score'].mean()
print("\nAverage marks in each degree stream:")
print(degree_avg_marks)

# f) To count the number of students in MBA
mba_count = df[df['Degree'] == 'MBA'].shape[0]
print("\nNumber of students in MBA:")
print(mba_count)

# g) To print the mode marks for BCA
bca_mode = df[df['Degree'] == 'BCA']['Score'].mode()
print("\nMode marks for BCA:")
print(bca_mode)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

categories = ['A', 'B', 'C', 'D']
values = [3, 7, 8, 5]
x = np.arange(1, 11)
y = np.random.randint(1, 10, 10)

# Bar chart
plt.figure(figsize=(15,12))
plt.subplot(3,3,1)
plt.bar(categories, values, label='Values')
plt.title("Bar chart")
plt.legend()

# Scatter Plot
plt.subplot(3,3,2)
plt.scatter(x, y, label='Scatter x vs y')
plt.scatter(y, x, c=y, cmap='rainbow', label='Scatter y vs x')
plt.scatter(categories, values, c=values, cmap='rainbow', label='Scatter categories vs values')
plt.title("Scatter Plot")
plt.legend()

# Line Chart
plt.subplot(3,3,3)
plt.plot(x, y, label='Line x vs y')
plt.title("Line Chart")
plt.legend()

# Area Chart
plt.subplot(3,3,4)
plt.fill_between(x, y, color='skyblue', alpha=0.2, label='Area under curve')
plt.plot(x, y, color="slateblue", alpha=0.6, label='Line plot')
plt.title("Area Chart")
plt.legend()

# Pie Chart
plt.subplot(3,3,5)
plt.pie(values, labels=categories, autopct="%.2f%%")
plt.title("Pie Chart")
plt.legend(loc="best")

# Histogram
plt.subplot(3,3,6)
plt.hist(y, bins=5, label='Histogram')
plt.title("Histogram")
plt.legend()

# Cumulative Distribution (Ogive)
plt.subplot(3,3,7)
plt.hist(y, bins=5, cumulative=True, histtype='step', color='red', label='Ogive')
plt.title("Ogive")
plt.legend()

# Dot plot
plt.subplot(3,3,8)
plt.plot(x, y, '*', markersize=7, label='Dot plot')
plt.title("Dot Plot")
plt.legend()

# Stem and leaf display
plt.subplot(3,3,9)
plt.stem(x, y, label='Stem plot')
plt.title("Stem-and-Leaf")
plt.legend()

plt.tight_layout()
plt.show()
```

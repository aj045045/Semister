```python
arr1 = np.array([12,3,4,5,6,7,8,9,10])
mean = arr1.mean()
std = arr1.std()
print("mean: ",mean,"std: ",std)
p_value = stats.ttest_1samp(arr1,popmean=1)
print(p_value)
# t_tests =
```

```python
arr2 = np.array([[45,4,32],[4,23,6],[45,23,9]])
print("Determinent ",linalg.det(arr2))
print("\nInverse value",linalg.inv(arr2))
print("\nEigan value",linalg.eig(arr2))
```

```python
import pandas as pd

df = pd.read_csv('./dataset/students_scores.csv')
print(pd.DataFrame(df))
print("\n Math Means",df['Math'].mean())
print("\n English Median",df['English'].median())
```

```python
print("\n Sort by Math \n",df.sort_values(by='Math',ascending=True))
print("\n Sort by Math and English \n",df.sort_values(by=['English','Math'],ascending=[True,False]))
df['Average'] = df[['Math','English','Science','History']].mean(axis=1)
print("\nAdd column:\n",df)
```

```python
df = df.drop(columns=['Average'])
print("\nDrop Columns\n",df)
print(df.groupby('Student').mean())
```

```python
data_with_nan = {
    "Student": ["Aansh", "Ankit", "Tirth", "Tej", "Hiren"],
    "Math": [85, 91, 78, None, 90],
    "English": [78, None, 95, 76, 85],
    "Science": [92, 89, None, 85, 91],
    "History": [74, 77, 80, 85, None]
}

df_nan = pd.DataFrame(data_with_nan)

print("\nDataFrame with missing values:")
print(df_nan)

print("\nFill missing values with 0:")
df_filled = df_nan.fillna(0)
print(df_filled)

print("\nDrop rows with missing values:")
df_dropped = df_nan['English'].dropna()
print(df_dropped)

print("\nFill missing values with the mean of each column:")
df_filled_mean = df_nan.fillna(df_nan['History'].mean())
print(df_filled_mean)
```

```python
# 9. Merging DataFrames
data_extra = {
    "Student": ["Aansh", "Ankit", "Tirth", "Tej", "Hiren"],
    "Art": [88, 92, 85, 87, 90]
}

df_extra = pd.DataFrame(data_extra)

print("\nMerging the original DataFrame with an additional DataFrame:")
df_merged = pd.merge(df, df_extra, on="Student")
print(df_merged)

# Pivot Tables
print("\nCreating a pivot table with the average scores:")
pivot = df.pivot_table(index="Student", values=["Math", "English", "Science", "History"], aggfunc="mean")
print(pivot)
```

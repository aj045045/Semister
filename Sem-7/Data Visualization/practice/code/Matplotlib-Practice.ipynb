## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("./dataset/student.csv")
df.head()
```

```python

python = df[df['Subject'] == 'Python']['Subject'].count()
wad = df[df['Subject'] == 'WAD']['Subject'].count()
networking = df[df['Subject'] == 'Networking']['Subject'].count()

plt.title("Student Subject")
plt.bar(['Python','WAD','Networking'],[python,wad,networking],label="Percentage",color="blue")
plt.xlabel("Subjects")
plt.ylabel("No of Student")
plt.legend()
plt.show()
```

```python
student_100_75 = df[(df.Percentage > 75 ) & (df.Percentage < 100)]['Percentage'].count()
student_75_50 = df[(df.Percentage > 50 ) & (df.Percentage < 75)]['Percentage'].count() 
student_50_25 = df[(df.Percentage > 25 ) & (df.Percentage < 50)]['Percentage'].count()
student_25 = df[df.Percentage < 25]['Percentage'].count() 

plt.title("Student Result")
plt.bar(['75 to 100','50 to 75'],[student_100_75,student_75_50],label="Pass",color="green")
plt.bar(['25 to 50','less than 25'],[student_50_25,student_25],color="red",label="Fail")
plt.xlabel("Student Percentage")
plt.ylabel("No. of student")
plt.legend()
plt.show()
```

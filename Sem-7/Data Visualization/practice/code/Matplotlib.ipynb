```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./dataset/student.csv")
df.head()
```

```python
# Bar Diagram

x = df['SID']
y = df['Percentage']
plt.bar(x,df['Percentage_Past'],label="Percentage_Past",color='red')
plt.bar(x,y,label="Percentage",color='blue')
plt.xlabel("Student ID")
plt.ylabel("Student Information")
plt.title("Bar Diagram")
plt.legend(loc='upper left')
plt.show()
```

```python

wad = df[df['Subject'] == 'WAD']['Subject'].value_counts()['WAD']
network= df[df['Subject'] == 'Networking']['Subject'].value_counts()['Networking']
python = df[df['Subject'] == 'Python']['Subject'].value_counts()['Python']

plt.bar(['WAD','Python','Networking'],[wad,python,network],label="Subject",color="green")
plt.ylabel("No of Student")
plt.xlabel("Subjects")
plt.title("Subject Bar Chart")
plt.legend()
plt.show()
```

```python
# Bar Diagram
r = df[(df.Percentage > 50)&(df.Percentage < 70)]
plt.bar(r['SID'],r['Percentage'],label="Percentage",color='blue')
plt.xlabel("Student ID")
plt.ylabel("Student Information")
plt.title("Bar Diagram")
plt.legend(loc='upper left')
plt.show()
```

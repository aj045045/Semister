## Exercise 1. Rectangle class:
- Write a Rectangle class in Python language, allowing you to build a rectangle with length and width attributes.
- Create a Perimeter() method to calculate the perimeter of the rectangle and a Area() method to calculate the area of the rectangle.
- Create a method display() that display the length, width, perimeter and area of an object created using an instantiation on rectangle class.
- Create a Parallelepipede child class inheriting from the Rectangle class and with a height attribute and another Volume() method to calculate the volume of the Parallelepiped.

```python
class Rectangle:
    
    def __init__(self,width,length):
        self._width = width
        self._length = length
    
    def perimeter(self):
        self._perimeter = 2*(self._length + self._width)
        return self._perimeter
    
    def area(self):
        self._area = self._length * self._width
        return self._area
    
    def display(self):
        print("The Width of Rectangle is :",self._width)
        print("The length of Rectangle is :",self._length)
        print("The Area of Rectangle is :",self._area)
        print("The Perimeter of Rectangle is :",self._perimeter)

class Parallelepiped(Rectangle):
    
    def __init__(self,height, width, length):
        self.height = height
        super().__init__(width,length)
    
    def volume(self):
        return self.height * self._length

box = Parallelepiped(10,20,30)
print(f"The Perimeter of a box is {box.perimeter()}")
print(f"The Perimeter of a box is {box.area()}")
box.display()
print(f"The Perimeter of a box is {box.volume()}")
```

## Exercise 2: Person class and child Student class
- Create a Python class Person with attributes: name and age of type string.
- Create a display() method that displays the name and age of an object created via the Person class.
- Create a child class Student which inherits from the Person class and which also has a section attribute.
- Create a method displayStudent() that displays the name, age and section of an object created via the Student class.
- Create a student object via an instantiation on the Student class and then test the displayStudent method.
 

```python
class Person:
    def __init__(self, name, age):
        self._name = name
        self._age = age
        
    def displayPerson(self):
        print(f"The name of a person is \"{self._name}\" and his age is \"{self._age}\"")

class Student(Person):
    def __init__(self, name, age, section):
        self.__section = section
        super().__init__(name, age)
    
    def displayStudent(self):
        print(f"The section of a Student is \"{self.__section}\"")

name = str(input("Enter the name :"))
age = str(input("Enter the age :"))
section = str(input("Enter the section :"))
anshJYadav = Student(name,age,section)
anshJYadav.displayPerson()
anshJYadav.displayStudent()
```

## Exercise 3 Class Book
- Define a Book class with the following attributes: Title, Author (Full name), Price.
- Define a constructor used to initialize the attributes of the method with values entered by the user.
- Set the View() method to display information for the current book.
- Write a program to testing the Book class.

```python
class Book:
    def __init__(self, title, author, price):
        self.__title = title
        self.__author = author
        self.__price = price

    def view(self):
        print(f"\nThe title of a book is \"{self.__title}\" by the author \"{self.__author}\" you can get this book for just rupees \"{self.__price}\"")
        
listOfBooks = []
numOfBooks = int(input("Enter the number of books :"))
for x in range(numOfBooks):
    print(f"\n========[ ENTER THE {x+1} BOOKS DETAILS ]==========")
    titleOfBook = str(input("Enter the title of the book:"))
    authorOfBook = str(input("Enter the author of the book:"))
    priceOfBook = str(input("Enter the price of the book:"))
    listOfBooks.append(Book(titleOfBook, authorOfBook, priceOfBook))

for x in range(numOfBooks):
    listOfBooks[x].view()
```

## Exercise 5: Write a Python class BankAccount 
- Attributes like account_number, balance, date_of_opening and customer_name, and methods like deposit, withdraw, and check_balance.

```python
class BankAccount:
    
    def __init__(self, accountNumber, customerName, date, balance ):
        self.__accountNumber = accountNumber
        self.__balance = balance
        self.__dateOfOpening = date
        self.__customerName = customerName
        
    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        self.__balance -= amount
    
    def checkBalance(self):
        print(f"\nThe account number of customer \"{self.__customerName}\" is {self.__accountNumber} on \"{self.__dateOfOpening}\" having balance of rupees {self.__balance}")
listOfUser = {}
listOfName =[]
numOfCustomers = int(input("Enter the number of Customer to create :"))
for x in range(numOfCustomers):
    accountName = str(input("Enter the account Customer Name :"))
    accountNumber = int(input("Enter the account Number :"))
    accountDate = str(input("Enter the account Date of opening :"))
    accountBalance = int(input("Enter the account opening Balance :"))
    listOfUser[accountName] = (BankAccount(accountNumber, accountName, accountDate, accountBalance))
    listOfName.append(accountName)
print(listOfUser)
print(listOfName)

x = input("Enter name to change value ")
if x in listOfName:
    deposit = int(input("Enter the deposit amount :"))
    listOfUser[x].deposit(deposit)
    withdraw = int(input("Enter the withdraw amount :"))
    listOfUser[x].withdraw(withdraw)
    listOfUser[x].checkBalance()
```

## Exercise 6:Write a Python class to reverse a string word by word.
- Input string : 'hello' 
- Expected Output : 'olleh'

```python
class String:
    def __init__(self):
        self.__stringInput = ""
        
    def inputString(self, string):
        self.__stringInput = string
    
    def reverseString(self):
        print(f"\nBefore reversing a string : {self.__stringInput}")
        self.__stringInput = self.__stringInput[::-1]
        print(f"After reversing a string : {self.__stringInput}")

tempString = str(input("Enter the string to reverse :"))
stringObject = String()
stringObject.inputString(tempString)
stringObject.reverseString()
```

# Exercise 8: Write a program to create a child class Teacher that will inherit the properties of Parent class Staff

```python
class Staff:
    
    def __init__(self, department, hod, staffMember):
        self.__departmentName = department
        self.__headOfDepartment = hod
        self._numberOfStaff = staffMember
    
    def displayStaff(self):
        print("\n-------------[ DEPARTMENET DETAILS ]----------------")
        print("Department Name :",self.__departmentName)
        print("Department Head of Director :",self.__headOfDepartment)
        print("Number of Staff member :",self._numberOfStaff)
    
class Teacher(Staff):
    
    def __init__(self, department, hod, staffMember):
        self.__facultyMember = []
        super().__init__(department, hod, staffMember)
    
    def addFaculty(self, facultyMember):
        self.__facultyMember = facultyMember
    
    def displayDetails(self):
        super().displayStaff()
        print("Displaying faculty details")
        for fac in range(self._numberOfStaff):
            print(f"- {self.__facultyMember[fac]}")

dep = str(input("Enter the Department name :"))
hod = str(input("Enter the Head of department name :"))
numStaff = int(input("Enter number of staff member :"))
dcs = Teacher(dep, hod, numStaff)
staffName = []

for name in range(numStaff):
    memberName = str(input("Enter the faculty name :"))
    staffName.append(memberName)
    
dcs.addFaculty(staffName)
dcs.displayDetails()
```

# Exercise 9: Write a Python program that checks if one class is a subclass of another.

```python
class Vehicles:
 
    def __init__(vehicleType):
        print('Vehicles is a ', vehicleType)

class Car(Vehicles):
 
    def __init__(self):
        Vehicles.__init__('Car')

class Submarine:
    
    def __init__(self):
        Vehicles.__init__('Sports Car')

print("The Vehicle has subclass Car :",issubclass(Car, Vehicles))
print("The Vehicle has subclass Sports Car :",issubclass(Submarine, Vehicles))
```

# Exercise 10: Write a Python class Inventory
- Attributes like item_id, item_name, stock_count, and price, and methods like add_item, update_item, and check_item_details.

```python
class Inventory:
    def __init__(self):
        self.items = {}

    def add_item(self, item_id, item_name, stock_count, price):
        if item_id not in self.items:
            self.items[item_id] = {
                "item_name": item_name,
                "stock_count": stock_count,
                "price": price,
            }
            print(f"Added item {item_name} with ID {item_id} to inventory.")
        else:
            print(f"Item with ID {item_id} already exists in inventory.")

    def update_item(self, item_id, stock_count=None, price=None):
        if item_id in self.items:
            if stock_count is not None:
                self.items[item_id]["stock_count"] = stock_count
            if price is not None:
                self.items[item_id]["price"] = price
            print(f"Updated item with ID {item_id} in inventory.")
        else:
            print(f"Item with ID {item_id} does not exist in inventory.")

    def check_item_details(self, item_id):
        if item_id in self.items:
            item = self.items[item_id]
            print(f"Item ID: {item_id}")
            print(f"Item Name: {item['item_name']}")
            print(f"Stock Count: {item['stock_count']}")
            print(f"Price: {item['price']}")
        else:
            print(f"Item with ID {item_id} does not exist in inventory.")

inventory = Inventory()

inventory.add_item("A001", "Item A", 10, 19.99)
inventory.add_item("B001", "Item B", 5, 29.99)
inventory.update_item("A001", stock_count=15)
inventory.check_item_details("B001")
```
